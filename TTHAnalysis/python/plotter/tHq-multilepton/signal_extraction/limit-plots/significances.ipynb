```python
import pandas as pd
import numpy as np
from functools import partial
from ctcv_helper import print_table
from ctcv_helper import read_dataframe
from ctcv_helper import read_dataframe_ratio
```

```python
# Read csv files
df_sigs = pd.DataFrame.from_csv("significance_obs_K6_cv_0p5.csv", sep=",", index_col=None)
df_sigs = df_sigs.append(pd.DataFrame.from_csv("significance_obs_K6_cv_1p0.csv", sep=",", index_col=None), ignore_index=True)
df_sigs = df_sigs.append(pd.DataFrame.from_csv("significance_obs_K6_cv_1p5.csv", sep=",", index_col=None), ignore_index=True)
df_exp = pd.DataFrame.from_csv("significance_exp_K6_cv_0p5.csv", sep=",", index_col=None)
df_exp = df_exp.append(pd.DataFrame.from_csv("significance_exp_K6_cv_1p0.csv", sep=",", index_col=None), ignore_index=True)
df_exp = df_exp.append(pd.DataFrame.from_csv("significance_exp_K6_cv_1p5.csv", sep=",", index_col=None), ignore_index=True)
```

```python
# Print significances
print " alpha  Ct/CV    CV=0.5     CV=1.0     CV=1.5"
print_table([partial(read_dataframe, df=df_sigs, att='significance')],
             linepat=" %6.3f")
```

```python
df_sigs_new = pd.DataFrame()
df_sigs_new['alpha'] = np.sign(df_sigs.cf*df_sigs.cv)*df_sigs.cf**2/(df_sigs.cf**2+df_sigs.cv**2)
df_sigs_new['alpha'] = df_sigs_new.alpha.round(4)
df_sigs_new['ratio'] = df_sigs.cf/df_sigs.cv
df_sigs_new['ratio'] = df_sigs_new.ratio.round(3)
df_sigs_new['observed'] = df_sigs.significance
df_sigs_new['twosigdown'] = df_exp.twosigdown
df_sigs_new['onesigdown'] = df_exp.onesigdown
df_sigs_new['expected'] = df_exp.exp
df_sigs_new['onesigup'] = df_exp.onesigup
df_sigs_new['twosigup'] = df_exp.twosigup

# Remove the duplicate entries, sort, reset the index, and write to csv
df_sigs_new.drop_duplicates(subset='ratio', inplace=True)
df_sigs_new.sort_values(by='ratio', inplace=True)
df_sigs_new.index = range(1,len(df_sigs_new)+1)
df_sigs_new.to_csv("significances.csv")
print "Wrote to significances.csv"
```
